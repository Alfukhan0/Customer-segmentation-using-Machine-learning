```python

# ====================================================================
#  Customer Segmentation – Full Clustering + PCA Analysis
#  K-Means | Hierarchical | DBSCAN | Fuzzy C-Means  →  PCA on Best
#  NOTE: matplotlib Agg backend used – all figures saved as PNG files
# ====================================================================

import matplotlib
matplotlib.use("Agg")           # ← non-interactive: no blocking windows

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import (silhouette_score, davies_bouldin_score,
                             calinski_harabasz_score)
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from scipy.cluster.hierarchy import dendrogram, linkage

try:
    import skfuzzy as fuzz
    FUZZY_AVAILABLE = True
    print("  [OK] scikit-fuzzy loaded.")
except ImportError:
    FUZZY_AVAILABLE = False
    print("  [WARN] scikit-fuzzy not found – Fuzzy C-Means skipped.")

PALETTE = ["#4C72B0","#DD8452","#55A868","#C44E52",
           "#8172B2","#937860","#DA8BC3","#8C8C8C"]

# ─────────────────────────────────────────────
# 1. LOAD & PREPROCESS
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  STEP 1 – Loading & Preprocessing")
print("="*60)

df   = pd.read_csv("Extended_Mall_Customers_Clean.csv")
print(f"  Shape  : {df.shape}")
print(f"  Columns: {list(df.columns)}")

le     = LabelEncoder()
df_enc = df.copy()
for col in ["Gender", "Profession", "Membership"]:
    df_enc[col] = le.fit_transform(df_enc[col].astype(str))

features = df_enc.drop(columns=["CustomerID"])
scaler   = StandardScaler()
X        = scaler.fit_transform(features)
print(f"  Scaled feature matrix: {X.shape}")


def eval_metrics(X, labels, name=""):
    n_uniq = len(set(labels)) - (1 if -1 in labels else 0)
    if n_uniq < 2:
        print(f"  [{name}] Only {n_uniq} cluster – scores skipped.")
        return None, None, None
    sil = silhouette_score(X, labels)
    db  = davies_bouldin_score(X, labels)
    ch  = calinski_harabasz_score(X, labels)
    print(f"  [{name}] Silhouette={sil:.4f}  DB={db:.4f}  CH={ch:.1f}")
    return sil, db, ch


# ─────────────────────────────────────────────
# 2. K-MEANS  – Elbow + Silhouette
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  STEP 2 – K-Means")
print("="*60)

K_RANGE   = range(2, 11)
inertias  = []
sil_km    = []

for k in K_RANGE:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)
    sil_km.append(silhouette_score(X, km.labels_))

best_k = list(K_RANGE)[np.argmax(sil_km)]
print(f"  Best K = {best_k}  Silhouette = {max(sil_km):.4f}")

km_best   = KMeans(n_clusters=best_k, random_state=42, n_init=10)
km_labels = km_best.fit_predict(X)
km_sil, km_db, km_ch = eval_metrics(X, km_labels, "K-Means")

# --- Elbow + Silhouette figure
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("K-Means – Elbow & Silhouette", fontsize=15, fontweight="bold")

axes[0].plot(list(K_RANGE), inertias, "o-", color=PALETTE[0], lw=2, ms=8)
axes[0].axvline(best_k, ls="--", color="red", label=f"Best k={best_k}")
axes[0].set_xlabel("k"); axes[0].set_ylabel("Inertia (WCSS)")
axes[0].set_title("Elbow Method"); axes[0].legend()

axes[1].plot(list(K_RANGE), sil_km, "s-", color=PALETTE[1], lw=2, ms=8)
axes[1].axvline(best_k, ls="--", color="red", label=f"Best k={best_k}")
axes[1].set_xlabel("k"); axes[1].set_ylabel("Silhouette Score")
axes[1].set_title("Silhouette Scores"); axes[1].legend()

plt.tight_layout()
plt.savefig("kmeans_elbow_silhouette.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: kmeans_elbow_silhouette.png")


# ─────────────────────────────────────────────
# 3. HIERARCHICAL – Dendrogram + sweep
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  STEP 3 – Hierarchical Clustering")
print("="*60)

np.random.seed(42)
idx      = np.random.choice(len(X), min(150, len(X)), replace=False)
linked   = linkage(X[idx], method="ward")

fig, ax = plt.subplots(figsize=(16, 6))
dendrogram(linked, ax=ax, truncate_mode="lastp", p=30,
           color_threshold=0.7*max(linked[:,2]),
           leaf_rotation=90, leaf_font_size=9)
ax.set_title("Hierarchical – Dendrogram (Ward)", fontsize=14, fontweight="bold")
ax.set_xlabel("Samples / Cluster"); ax.set_ylabel("Distance")
plt.tight_layout()
plt.savefig("hierarchical_dendrogram.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: hierarchical_dendrogram.png")

sil_hc = []
for k in K_RANGE:
    hc  = AgglomerativeClustering(n_clusters=k, linkage="ward")
    lbl = hc.fit_predict(X)
    sil_hc.append(silhouette_score(X, lbl))

best_k_hc = list(K_RANGE)[np.argmax(sil_hc)]
print(f"  Best K = {best_k_hc}  Silhouette = {max(sil_hc):.4f}")

hc_best   = AgglomerativeClustering(n_clusters=best_k_hc, linkage="ward")
hc_labels = hc_best.fit_predict(X)
hc_sil, hc_db, hc_ch = eval_metrics(X, hc_labels, "Hierarchical")

# Silhouette sweep plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(list(K_RANGE), sil_hc, "^-", color=PALETTE[2], lw=2, ms=8)
ax.axvline(best_k_hc, ls="--", color="red", label=f"Best k={best_k_hc}")
ax.set_title("Hierarchical – Silhouette vs k", fontsize=13, fontweight="bold")
ax.set_xlabel("k"); ax.set_ylabel("Silhouette Score"); ax.legend()
plt.tight_layout()
plt.savefig("hierarchical_silhouette.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: hierarchical_silhouette.png")


# ─────────────────────────────────────────────
# 4. DBSCAN – k-distance + param search
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  STEP 4 – DBSCAN")
print("="*60)

nbrs      = NearestNeighbors(n_neighbors=5).fit(X)
dists, _  = nbrs.kneighbors(X)
k_dists   = np.sort(dists[:, -1])

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(k_dists, color=PALETTE[2], lw=2)
ax.set_title("DBSCAN – k-Distance Graph (k=5)", fontsize=13, fontweight="bold")
ax.set_xlabel("Points (sorted)"); ax.set_ylabel("5-NN Distance")
plt.tight_layout()
plt.savefig("dbscan_kdistance.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: dbscan_kdistance.png")

best_sil_db, best_eps, best_min = -1, 0.5, 5
db_results = []

for eps in np.arange(0.3, 2.1, 0.2):
    for min_s in [3, 5, 8, 10]:
        db  = DBSCAN(eps=round(eps,1), min_samples=min_s)
        lbl = db.fit_predict(X)
        n_c = len(set(lbl)) - (1 if -1 in lbl else 0)
        noi = list(lbl).count(-1)
        if n_c >= 2:
            s = silhouette_score(X, lbl)
            db_results.append((round(eps,1), min_s, n_c, noi, s))
            if s > best_sil_db:
                best_sil_db, best_eps, best_min = s, round(eps,1), min_s

dbscan_best = DBSCAN(eps=best_eps, min_samples=best_min)
db_labels   = dbscan_best.fit_predict(X)
n_db_cl     = len(set(db_labels)) - (1 if -1 in db_labels else 0)
noise_pts   = list(db_labels).count(-1)
print(f"  Best eps={best_eps}  min_samples={best_min}")
print(f"  Clusters={n_db_cl}  Noise={noise_pts}")
db_sil, db_db, db_ch = eval_metrics(X, db_labels, "DBSCAN")

db_df = pd.DataFrame(db_results,
        columns=["eps","min_samples","clusters","noise","silhouette"])
print("\n  Top-10 DBSCAN param combos:")
print(db_df.sort_values("silhouette",ascending=False).head(10).to_string(index=False))


# ─────────────────────────────────────────────
# 5. FUZZY C-MEANS
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  STEP 5 – Fuzzy C-Means")
print("="*60)

fcm_sil = fcm_db = fcm_ch = fcm_labels = best_c_fcm = None

if FUZZY_AVAILABLE:
    X_T, c_range = X.T, range(2, 11)
    sil_fcm = []

    for c in c_range:
        _, u, *_ = fuzz.cluster.cmeans(
            X_T, c=c, m=2, error=0.005, maxiter=500, seed=42)
        lbl = np.argmax(u, axis=0)
        sil_fcm.append(silhouette_score(X, lbl))
        print(f"    c={c}  sil={sil_fcm[-1]:.4f}")

    best_c_fcm = list(c_range)[np.argmax(sil_fcm)]
    print(f"  Best C = {best_c_fcm}  Silhouette = {max(sil_fcm):.4f}")

    _, u_best, *_ = fuzz.cluster.cmeans(
        X_T, c=best_c_fcm, m=2, error=0.005, maxiter=500, seed=42)
    fcm_labels = np.argmax(u_best, axis=0)
    fcm_sil, fcm_db, fcm_ch = eval_metrics(X, fcm_labels, "Fuzzy C-Means")

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(list(c_range), sil_fcm, "D-", color=PALETTE[3], lw=2, ms=8)
    ax.axvline(best_c_fcm, ls="--", color="red", label=f"Best C={best_c_fcm}")
    ax.set_title("Fuzzy C-Means – Silhouette vs C", fontsize=13, fontweight="bold")
    ax.set_xlabel("C"); ax.set_ylabel("Silhouette Score"); ax.legend()
    plt.tight_layout()
    plt.savefig("fcm_silhouette.png", dpi=150, bbox_inches="tight")
    plt.close()
    print("  Saved: fcm_silhouette.png")
else:
    print("  SKIPPED (scikit-fuzzy not installed).")


# ─────────────────────────────────────────────
# 6. COMPARISON TABLE + BAR CHART
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  STEP 6 – Algorithm Comparison")
print("="*60)

results = {
    "Algorithm"        : ["K-Means", "Hierarchical", "DBSCAN", "Fuzzy C-Means"],
    "Clusters"         : [best_k, best_k_hc, n_db_cl,
                          best_c_fcm if FUZZY_AVAILABLE else "N/A"],
    "Silhouette (↑)"   : [km_sil,  hc_sil,  db_sil,  fcm_sil],
    "Davies-Bouldin (↓)": [km_db,  hc_db,   db_db,   fcm_db],
    "Calinski-H (↑)"   : [km_ch,   hc_ch,   db_ch,   fcm_ch],
}
res_df = pd.DataFrame(results)
print(res_df.to_string(index=False))

# Best by silhouette
scores    = {"K-Means": km_sil, "Hierarchical": hc_sil,
             "DBSCAN": db_sil, "Fuzzy C-Means": fcm_sil}
valid     = {k: v for k, v in scores.items() if v is not None}
best_algo = max(valid, key=valid.get)
print(f"\n  🏆 BEST: {best_algo}  (Silhouette = {valid[best_algo]:.4f})")

algo_labels_map = {
    "K-Means"      : km_labels,
    "Hierarchical" : hc_labels,
    "DBSCAN"       : db_labels,
    "Fuzzy C-Means": fcm_labels,
}
best_labels = algo_labels_map[best_algo]

# Bar comparison
metrics_keys = ["Silhouette (↑)", "Davies-Bouldin (↓)", "Calinski-H (↑)"]
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Algorithm Comparison – Evaluation Metrics", fontsize=14, fontweight="bold")

for ax, mk in zip(axes, metrics_keys):
    vals   = [v if v is not None else 0 for v in results[mk]]
    names  = results["Algorithm"]
    bars   = ax.bar(names, vals, color=PALETTE[:4], edgecolor="white", lw=1.5)
    ax.set_title(mk, fontsize=11)
    ax.set_xticklabels(names, rotation=15, ha="right")
    mx = max(vals) if max(vals) > 0 else 1
    for bar, val in zip(bars, vals):
        if val:
            ax.text(bar.get_x()+bar.get_width()/2,
                    bar.get_height()+0.01*mx,
                    f"{val:.3f}", ha="center", va="bottom", fontsize=8)

plt.tight_layout()
plt.savefig("algorithm_comparison.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: algorithm_comparison.png")


# ─────────────────────────────────────────────
# 7. PCA ON BEST ALGORITHM
# ─────────────────────────────────────────────
print("\n" + "="*60)
print(f"  STEP 7 – PCA on Best ({best_algo})")
print("="*60)

pca2     = PCA(n_components=2, random_state=42)
X_2d     = pca2.fit_transform(X)

pca3     = PCA(n_components=3, random_state=42)
X_3d     = pca3.fit_transform(X)

pca_full = PCA(random_state=42).fit(X)
cum_var  = np.cumsum(pca_full.explained_variance_ratio_)

print(f"  2-D explains {pca2.explained_variance_ratio_.sum()*100:.1f}% variance")
print(f"  3-D explains {pca3.explained_variance_ratio_.sum()*100:.1f}% variance")

# --- 7a. Explained variance
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("PCA – Explained Variance", fontsize=14, fontweight="bold")

axes[0].bar(range(1, len(pca_full.explained_variance_ratio_)+1),
            pca_full.explained_variance_ratio_*100,
            color=PALETTE[0], edgecolor="white")
axes[0].set_title("Individual Variance per Component")
axes[0].set_xlabel("Principal Component"); axes[0].set_ylabel("Variance %")

axes[1].plot(range(1, len(cum_var)+1), cum_var*100,
             "o-", color=PALETTE[1], lw=2, ms=6)
axes[1].axhline(95, ls="--", color="red", label="95% threshold")
axes[1].set_title("Cumulative Explained Variance")
axes[1].set_xlabel("Components"); axes[1].set_ylabel("Cumulative %")
axes[1].legend()

plt.tight_layout()
plt.savefig("pca_variance.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: pca_variance.png")

# --- 7b. 2-D PCA scatter
unique_lbl   = sorted(set(best_labels))
n_cl_best    = len([l for l in unique_lbl if l != -1])
cmap_c       = plt.cm.get_cmap("tab10", n_cl_best+1)

fig, ax = plt.subplots(figsize=(10, 7))
for lbl in unique_lbl:
    mask  = best_labels == lbl
    color = "grey" if lbl == -1 else cmap_c(lbl)
    label = "Noise" if lbl == -1 else f"Cluster {lbl}"
    ax.scatter(X_2d[mask,0], X_2d[mask,1],
               c=[color]*mask.sum(), s=60, alpha=0.75,
               label=label, edgecolors="white", lw=0.5)

ax.set_title(f"PCA 2D – {best_algo}", fontsize=14, fontweight="bold")
ax.set_xlabel(f"PC1 ({pca2.explained_variance_ratio_[0]*100:.1f}%)")
ax.set_ylabel(f"PC2 ({pca2.explained_variance_ratio_[1]*100:.1f}%)")
ax.legend(loc="best", framealpha=0.7)
plt.tight_layout()
plt.savefig("pca_2d_clusters.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: pca_2d_clusters.png")

# --- 7c. 3-D PCA scatter
fig = plt.figure(figsize=(12, 8))
ax3 = fig.add_subplot(111, projection="3d")
for lbl in unique_lbl:
    mask  = best_labels == lbl
    color = "grey" if lbl == -1 else cmap_c(lbl)
    label = "Noise" if lbl == -1 else f"Cluster {lbl}"
    ax3.scatter(X_3d[mask,0], X_3d[mask,1], X_3d[mask,2],
                c=[color]*mask.sum(), s=50, alpha=0.7,
                label=label, edgecolors="white", lw=0.3)

ax3.set_title(f"PCA 3D – {best_algo}", fontsize=13, fontweight="bold")
ax3.set_xlabel(f"PC1 ({pca3.explained_variance_ratio_[0]*100:.1f}%)")
ax3.set_ylabel(f"PC2 ({pca3.explained_variance_ratio_[1]*100:.1f}%)")
ax3.set_zlabel(f"PC3 ({pca3.explained_variance_ratio_[2]*100:.1f}%)")
ax3.legend(loc="upper left")
plt.tight_layout()
plt.savefig("pca_3d_clusters.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: pca_3d_clusters.png")

# --- 7d. Biplot
fig, ax = plt.subplots(figsize=(10, 8))
sc = ax.scatter(X_2d[:,0], X_2d[:,1],
                c=best_labels, cmap="tab10", s=30, alpha=0.5)
loadings = pca2.components_.T
feat_names = list(features.columns)
for i, name in enumerate(feat_names):
    ax.arrow(0, 0, loadings[i,0]*3, loadings[i,1]*3,
             head_width=0.08, head_length=0.05,
             fc=PALETTE[4], ec=PALETTE[4], lw=1.5)
    ax.text(loadings[i,0]*3.4, loadings[i,1]*3.4,
            name, fontsize=8, color="darkred", ha="center")
plt.colorbar(sc, ax=ax, label="Cluster")
ax.set_title(f"PCA Biplot – {best_algo}", fontsize=13, fontweight="bold")
ax.set_xlabel(f"PC1 ({pca2.explained_variance_ratio_[0]*100:.1f}%)")
ax.set_ylabel(f"PC2 ({pca2.explained_variance_ratio_[1]*100:.1f}%)")
ax.axhline(0, ls="--", color="grey", lw=0.8)
ax.axvline(0, ls="--", color="grey", lw=0.8)
plt.tight_layout()
plt.savefig("pca_biplot.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: pca_biplot.png")

# --- 7e. Cluster profile heatmap
df_profile = features.copy()
df_profile["Cluster"] = best_labels
profile    = df_profile[df_profile["Cluster"] != -1].groupby("Cluster").mean()
profile_sc = (profile - profile.min()) / (profile.max() - profile.min() + 1e-9)

fig, ax = plt.subplots(figsize=(14, max(5, len(profile)*1.2)))
sns.heatmap(profile_sc, annot=True, fmt=".2f", cmap="YlOrRd",
            linewidths=0.5, ax=ax, cbar_kws={"label": "Normalised Mean"})
ax.set_title(f"Cluster Profile Heatmap – {best_algo}", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("cluster_profile_heatmap.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: cluster_profile_heatmap.png")

# Save labelled CSV
df_out          = df.copy()
df_out["Cluster"] = best_labels
df_out.to_csv("clustered_customers.csv", index=False)
print("  Saved: clustered_customers.csv")


# ─────────────────────────────────────────────
# 8. FINAL SUMMARY
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("  FINAL SUMMARY")
print("="*60)
print(res_df.to_string(index=False))
print(f"\n  🏆 Best Algorithm   : {best_algo}")
print(f"  📊 Clusters Found   : {n_cl_best}")
print(f"  📈 Silhouette Score : {valid[best_algo]:.4f}")

print("\n  Output files generated:")
output_files = [
    "kmeans_elbow_silhouette.png",
    "hierarchical_dendrogram.png",
    "hierarchical_silhouette.png",
    "dbscan_kdistance.png",
    "fcm_silhouette.png" if FUZZY_AVAILABLE else "(fcm_silhouette.png – skipped)",
    "algorithm_comparison.png",
    "pca_variance.png",
    "pca_2d_clusters.png",
    "pca_3d_clusters.png",
    "pca_biplot.png",
    "cluster_profile_heatmap.png",
    "clustered_customers.csv",
]
for f in output_files:
    print(f"    ✓ {f}")
print("="*60)
print("  DONE.")
```
